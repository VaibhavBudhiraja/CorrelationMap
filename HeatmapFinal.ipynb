```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
print("Welcome setup is complete")
```

### Reading the data from excel or CSV files


```python
df = pd.read_csv('Hdpe.csv')    # File loading for excel and csv 
print(df)
```

### Droping the unused coloumn and rows

```python
df.drop(columns=['Unnamed: 0'], inplace= True)
df.head()
```

### Basic correlation  plot

```python
sns.heatmap(df.corr())
```

### Well decorated corelation plot using heatmap in seaborn 

```python
mat = np.triu(df.corr())   # upper triangular matrix
#plt.figure(figsize=(9,9))
sns.heatmap(df.corr(), annot=True, linewidth=.7, cmap='bwr', vmin=-1, vmax=1, mask=mat, square=True,  fmt=".1f")
#plt.show()
```
